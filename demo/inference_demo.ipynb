```python
config_file = '../configs/hrnet/cascade_rcnn_hrnetv2p_w32_20e.py'
#config_file = '../configs/hrnet/fcos_hrnetv2p_w32_gn_1x_1gpu_oi.py'
# download the checkpoint from model zoo and put it in `checkpoints/`
checkpoint_file = '../checkpoints/cascade_rcnn_hrnetv2p_w32_20e_20190522-55bec4ee.pth'
#checkpoint_file = '../work_dirs/fcos_hrnetv2p_w32_gn_1x_1gpu_oi/latest.pth'
```

```python
# build the model from a config file and a checkpoint file
model = init_detector(config_file, checkpoint_file, device='cuda:0')
```
